**ANSWER:** No. The product is a GRD High Resolution (GRDH), not Full Resolution (GRDF). Supporting metadata fields:

- resolution = High — explicitly not "Full".

**QUESTION (10pts): What is the line of sight angle at the near range of the scene and at the far range of the scene? Which edge of the scene (east/west) will be more susceptible to foreshortening/layover effects, and which edge of the scene will be more susceptible to shadow effects?**

**ANSWER:** **ANSWER:** For Sentinel-1 IW mode, the incidence angle runs from **~29° at near range** to
**~46° at far range** (readable from the `incidence_near` / `incidence_far` fields in SNAP's
Abstracted Metadata).

Which edge is near/far depends on the orbit direction (`PASS` field in metadata):

| Orbit | Near range (~29°) | Far range (~46°) |
|---|---|---|
| **Ascending** | West edge | East edge |
| **Descending** | East edge | West edge |

- **Near range edge** (~29°, steeper look angle) → more susceptible to **foreshortening and
layover**, because radar-facing slopes get compressed or fold over at steep look angles.
- **Far range edge** (~46°, more grazing look angle) → more susceptible to **shadow**, because
the shallow angle allows ridges to block illumination of terrain behind them.

## 2. Helper functions

These functions keep the notebook readable and reduce repeated plotting code.


```python
def read_raster(path, band=1):
    """Read a single-band raster and return array + metadata.
    Nodata values are converted to NaN.
    """
    with rasterio.open(path) as src:
        arr = src.read(band).astype("float32")
        meta = src.meta.copy()
        bounds = src.bounds
        transform = src.transform
        crs = src.crs
        nodata = src.nodata

    if nodata is not None:
        arr[arr == nodata] = np.nan

    return arr, meta, bounds, transform, crs


def plot_image(arr, title="", vmin=None, vmax=None, cmap="gray", figsize=(8, 6)):
    plt.figure(figsize=figsize)
    plt.imshow(arr, cmap=cmap, vmin=vmin, vmax=vmax)
    plt.colorbar()
    plt.title(title)
    plt.axis("off")
    plt.show()


def plot_side_by_side(arr1, arr2, title1, title2, vmin=None, vmax=None, cmap="gray", figsize=(14, 6)):
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    axes[0].imshow(arr1, cmap=cmap, vmin=vmin, vmax=vmax)
    axes[0].set_title(title1)
    axes[0].axis("off")

    im2 = axes[1].imshow(arr2, cmap=cmap, vmin=vmin, vmax=vmax)
    axes[1].set_title(title2)
    axes[1].axis("off")

    fig.colorbar(im2, ax=axes.ravel().tolist(), shrink=0.8)
    plt.show()


def to_db(arr, eps=1e-6):
    """Convert linear power to dB if needed.
    If your exported rasters are already in dB, do not apply this.
    """
    return 10 * np.log10(np.maximum(arr, eps))


def robust_limits(arr, pmin=2, pmax=98):
    valid = arr[np.isfinite(arr)]
    return np.percentile(valid, pmin), np.percentile(valid, pmax)


def histogram(arr, bins=100, title="", xlabel="", logy=False):
    valid = arr[np.isfinite(arr)]
    plt.figure(figsize=(8, 4))
    plt.hist(valid, bins=bins)
    if logy:
        plt.yscale("log")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Pixel count")
    plt.grid(True, alpha=0.3)
    plt.xlim(xmin=0,xmax=200)
    plt.show()

def plot_transects(arr1, arr2, label1="Before TC", label2="After TC", axis=0, title="Central transect"):
    t1 = central_transect(arr1, axis=axis)
    t2 = central_transect(arr2, axis=axis)

    plt.figure(figsize=(10, 4))
    plt.plot(t1, label=label1)
    plt.plot(t2, label=label2)
    plt.title(title)
    plt.xlabel("Pixel index")
    plt.ylabel("Backscatter")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

def pick_points(image, title="Click 2 points, then close the figure", cmap="gray", vmin=None, vmax=None):
    """Display an image and let the user click two points.

    Returns:
        pts: list of two (x, y) integer pixel tuples
    """
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.imshow(image, cmap=cmap, vmin=vmin, vmax=vmax)
    ax.set_title(title)
    ax.set_xlabel("X pixel")
    ax.set_ylabel("Y pixel")

    pts = plt.ginput(2, timeout=-1)
    plt.close(fig)

    pts = [(int(round(x)), int(round(y))) for x, y in pts]
    return pts


def extract_transect(arr, p0, p1, n=500):
    """Extract values along a line between two pixel coordinates."""
    x0, y0 = p0
    x1, y1 = p1

    xs = np.linspace(x0, x1, n)
    ys = np.linspace(y0, y1, n)

    xs_i = np.clip(np.round(xs).astype(int), 0, arr.shape[1] - 1)
    ys_i = np.clip(np.round(ys).astype(int), 0, arr.shape[0] - 1)

    values = arr[ys_i, xs_i]
    distance = np.arange(n)

    return distance, values, xs_i, ys_i


def plot_transect_on_image(image, p0, p1, title="", cmap="gray", vmin=None, vmax=None):
    """Plot an image with the selected transect overlaid."""
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.imshow(image, cmap=cmap, vmin=vmin, vmax=vmax)
    ax.plot([p0[0], p1[0]], [p0[1], p1[1]], 'r-', linewidth=2)
    ax.plot(p0[0], p0[1], 'go', markersize=8, label='Start')
    ax.plot(p1[0], p1[1], 'bo', markersize=8, label='End')
    ax.set_title(title)
    ax.set_xlabel("X pixel")
    ax.set_ylabel("Y pixel")
    ax.legend()
    plt.show()


def robust_limits(arr, pmin=2, pmax=98):
    valid = arr[np.isfinite(arr)]
    return np.percentile(valid, pmin), np.percentile(valid, pmax)
```

**QUESTION (10pts): Why is the CRS (coordinate reference system) for the Pre-TC view "None" while the post-TC has "EPSG:4326"?**

**ANSWER:** The pre-TC image is still in **radar/slant-range geometry** — pixels are
indexed by radar timing (range delay and azimuth time), not by geographic coordinates. No map
projection has been applied yet, so rasterio finds no CRS and returns `None`. After
**terrain correction**, the image is orthorectified and projected onto the Earth's surface,
assigning real geographic coordinates to each pixel — hence **EPSG:4326** (WGS84 lat/lon).


**QUESTION (10pts): Why can't we overlay the Pre-TC image with the DEM? Give two reasons.**

**ANSWER:** Two reasons:

1. **Different coordinate systems.** The pre-TC image has no CRS (radar geometry), while
the DEM is in a geographic CRS. There is no shared spatial reference that would allow
pixel-to-pixel alignment.

2. **Different geometry.** The pre-TC image is geometrically distorted by topography
(foreshortening, layover, shadow) — a pixel's position in radar coordinates does *not*
correspond to the same ground location as the matching coordinate in the DEM. Mountain
features will be spatially displaced relative to their true positions, so overlaying the
two would produce a meaningless result even if the CRS mismatch were somehow resolved.

## 7. Histograms before and after terrain correction

A histogram is a blunt but useful summary. It can show whether the distribution shifts after terrain correction.

**QUESTION (10pts): How did the distribution of backscatter change after terrain correction? Explain why.**

**ANSWER:** After terrain correction, the distribution of backscatter values becomes
**narrower and more symmetric**, with the extreme high values (very bright pixels) reduced
and the overall histogram shifted toward a more uniform spread.

This happens for two reasons:

1. **Geometric redistribution of pixels.** Before TC, steep radar-facing slopes are
compressed into fewer pixels (foreshortening/layover), concentrating high backscatter
values, while shadow zones contribute many near-zero pixels. Terrain correction
resamples the image to true ground geometry, spreading those compressed bright pixels
over their correct surface area and removing or flagging shadow/layover pixels — this
flattens the extremes in both tails of the distribution.

2. **Radiometric normalisation of slope angle.** Slopes facing the radar receive more
energy per unit area than flat ground (the effective antenna footprint is smaller on a
tilted surface), producing artificially high backscatter. After TC, pixel values are
normalised to the true ground area (sigma-nought → gamma-nought convention), reducing
the brightness of radar-facing slopes and raising the relative brightness of
away-facing slopes, pulling the distribution closer to the range expected for the
underlying land cover.

## 10. Incidence angle

If you exported an incidence-angle raster, this is a good place to inspect it and compare it with brightness.

**QUESTION (10pts): Where are bright/dark patterns linked to local geometry rather, and where are bright/dark patterns linked to land cover.**

**ANSWER:** 

**Geometry-driven patterns:**
Bright and dark patterns that closely follow the topography visible in the DEM are
dominated by local geometry rather than land cover:

- **Bright areas on radar-facing slopes** (slopes oriented toward the sensor) are bright
because the local incidence angle is small — more radar energy is returned per unit
ground area. This is a purely geometric effect independent of what is actually growing
or lying on that slope.
- **Dark areas in shadow zones** (behind ridges, in deep valleys on the far-range side)
are dark because the radar beam never reaches them — again, purely geometric.
- **Foreshortening zones** (compressed bright bands at the base of steep slopes) are
also geometry-driven; their brightness comes from many ground pixels being squeezed
into few image pixels.

In general: if a bright or dark pattern spatially correlates with ridges, valley walls,
or slope aspect visible in the DEM, it is geometry-driven.

**Land-cover-driven patterns:**
Patterns that do **not** follow the terrain relief but instead correspond to recognisable
land-use boundaries are driven by surface properties:

- **Valley floors and agricultural fields** tend to show moderate, relatively uniform
backscatter that contrasts with surrounding slopes regardless of slope angle.
- **Built-up areas** (towns in the Dolomite valleys) appear bright due to double-bounce
and corner reflector effects from buildings, even on flat terrain.
- **Forests vs. bare rock** differ in backscatter due to volume scattering vs. specular
or diffuse surface scattering — these contrasts persist even on slopes with similar
incidence angles.

In general: if a boundary between bright and dark cuts across the topography rather than
following it, it is most likely a land-cover boundary.

peaks are displaced toward the sensor, valleys away from it. A east-west transect
therefore shows the greatest positional mismatch between the pre-TC and post-TC
images, and the most dramatic shift in backscatter values after correction. A
north-south transect shows little such displacement because azimuth distortions are
much smaller.

## Reflection question

**QUESTION (10pts): If a researcher downloaded a ready-made GRD product and never thought about terrain effects, what kinds of interpretation errors might they make in a mountainous study area?**

**ANSWER:** A researcher ignoring terrain effects in a mountainous area could make
several serious interpretation errors:

1. **Misidentifying land cover.** Bright returns on steep radar-facing slopes are caused
by geometry, not by the surface material. A researcher might incorrectly classify a
bare rock slope as a built-up area or dense forest simply because it happens to face
the sensor.

2. **Positional errors.** Without terrain correction, features (ridges, rivers, forest
edges) are displaced in the range direction relative to their true ground positions.
Overlaying the SAR image with a map or another dataset would produce systematic
misalignment, leading to wrong conclusions about where phenomena are occurring.

3. **Incorrect change detection.** If two images are acquired from slightly different
geometries (e.g., ascending vs. descending pass), geometry-driven brightness
differences would be mistaken for real changes in land cover or surface conditions
between acquisitions.

4. **Biased quantitative estimates.** Any backscatter-based quantity (biomass, soil
moisture, snow water equivalent) would be systematically overestimated on
radar-facing slopes and underestimated on away-facing slopes, because the raw GRD
backscatter is not normalised to true ground area.

5. **Ignoring shadow and layover zones.** Areas with no valid data (shadow) or
geometrically folded data (layover) would be treated as real signal, introducing
corrupted values into any statistical analysis or model trained on the imagery.
