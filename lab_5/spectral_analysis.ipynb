```python
import numpy as np
import matplotlib.pyplot as plt
import spectral.io.envi as envi
from spectral import imshow

# Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

```python
# Define the path to your header file
hdr_path = 'data/images/221000_Odra_HS_Blok_A_008_VS_join_atm.hdr'

# Open the ENVI image
# Note: This creates a memory-mapped object (SpyFile), it doesn't load the whole file into RAM yet.
img = envi.open(hdr_path)

# Display basic metadata
print(f"Data Cube Dimensions (Rows, Columns, Bands): {img.shape}")
print(f"Number of Spectral Bands: {img.shape[2]}")
print(f"Data interleave type: {img.metadata.get('interleave', 'Unknown')}")
```
